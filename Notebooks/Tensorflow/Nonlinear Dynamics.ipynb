#### Lorenz Attractor  

```python
# Model Parameters

rho = 28.0
sigma = 10.0
beta = 8.0/3.0
inits = [0, 2, 20]

# Lorenz equations - Proof of concept for Tensorflow ODE solving

def lorenz_equation(state, t):
    x, y, z = hp.tf.unstack(state)
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return hp.tf.stack([dx, dy, dz])

# Session Definition

sess = hp.tf.Session()
state, info = sess.run(hp.generate_odesolution(lorenz_equation, inits, n=5000))
x, y, z = state.T
hp.plt.plot(x, z)
```
